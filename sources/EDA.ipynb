## 1. Phân tích tương quan

```python
# Nhận xét:
# Độ tuổi xét duyệt khoản vay chủ yếu rơi vào những người từ 20 đến 30 tuổi
```

```python
# Tính số lượng người trong mỗi nhóm tuổi
age_counts = df_no_trans['person_age'].value_counts().sort_index()

# Vẽ biểu đồ
plt.bar(age_counts.index, age_counts.values)
plt.xlabel('Độ tuổi')
plt.ylabel('Số lượng người')
plt.title('Phân bố độ tuổi')
plt.grid(True)
plt.show()
```

```python
# Nhận xét:
# Nhóm không được phê duyệt:
# Khoảng giá trị: Phần lớn các khoản vay bị từ chối có số tiền vay tập trung ở mức thấp và trung bình.
# Các giá trị ngoại lệ: Có một số ít trường hợp có số tiền vay rất cao, nằm ngoài khoảng biến động thông thường của nhóm này. 
#                    Điều này có thể cho thấy một số yếu tố đặc biệt liên quan đến các khoản vay lớn bị từ chối.
# Độ phân tán: Khoảng cách giữa các quartile (các vạch của hộp) khá lớn, cho thấy sự phân tán của dữ liệu trong nhóm này tương đối cao.

# Nhóm được phê duyệt:
# Khoảng giá trị: Số tiền vay của nhóm này thường cao hơn so với nhóm không được phê duyệt.
```

```python
# Tạo biểu đồ box plot
plt.boxplot([df_no_trans[df_no_trans['loan_status'] == status]['loan_amnt'] for status in df_no_trans['loan_status'].unique()],
            labels=df_no_trans['loan_status'].unique())

# Thêm tiêu đề và nhãn cho các trục
plt.title('Biểu đồ Box Plot của Khoản Vay theo Tình Trạng')
plt.xlabel('Tình Trạng Khoản Vay')
plt.ylabel('Số Tiền Vay')

plt.show()
```

```python
# Tạo biểu đồ phân tán
plt.scatter(df_no_trans['loan_percent_income'], df_no_trans['loan_status'])

# Thêm tiêu đề và nhãn cho các trục
plt.title('Mối quan hệ giữa tỷ lệ khoản vay và tình trạng khoản vay')
plt.xlabel('Tỷ lệ khoản vay so với thu nhập')
plt.ylabel('Tình trạng khoản vay')

# Hiển thị biểu đồ
plt.show()
```

```python
# Nhận xét:
# Nhóm không được phê duyệt:
# Lãi suất trung bình: Nhóm này có xu hướng có lãi suất trung bình thấp hơn so với nhóm được phê duyệt.
# Độ phân tán: Khoảng cách giữa các quartile (các vạch của hộp) khá lớn, cho thấy sự phân tán của dữ liệu trong nhóm này tương đối cao. 
# Lãi suất vay của những người không được phê duyệt có sự khác biệt khá lớn.

# Nhóm được phê duyệt:
# Nhóm này có xu hướng có lãi suất trung bình cao hơn so với nhóm không được phê duyệt.
```

```python
# Tạo danh sách các giá trị lãi suất cho từng nhóm tình trạng khoản vay
loan_status_groups = df_no_trans.groupby('loan_status')['loan_int_rate']
loan_int_rates = [group.values for _, group in loan_status_groups]

# Tạo biểu đồ boxplot
plt.boxplot(loan_int_rates, labels=loan_status_groups.groups)

# Thêm tiêu đề và nhãn cho các trục
plt.title('So sánh lãi suất vay giữa các nhóm khoản vay')
plt.xlabel('Tình trạng khoản vay')
plt.ylabel('Lãi suất vay')

# Hiển thị biểu đồ
plt.show()
```

```python
# Nhận xét:
# Nhóm không được phê duyệt:
# Nhóm này có đỉnh phân phối ở khoảng thời gian làm việc ngắn hơn, tập trung nhiều ở những người có thời gian làm việc dưới 10 năm.

# Nhóm người được phê duyệt:
# Tương tự so với nhóm trên nhưng có xu hướng dịch sang phải một ít, cho thấy thời gian làm việc có nhiều hơn một vài năm

# Chung: Nhìn chung, những người có thời gian làm việc dài hơn có xu hướng được phê duyệt khoản vay nhiều hơn.
```

```python
# Tạo histogram cho từng nhóm loan_status
for status in df_no_trans['loan_status'].unique():
    plt.hist(df_no_trans[df_no_trans['loan_status'] == status]['person_emp_length'], label=status, alpha=0.5)

plt.legend()
plt.xlabel('Thời gian làm việc (năm)')
plt.ylabel('Số lượng')
plt.title('Phân phối thời gian làm việc theo tình trạng khoản vay')

plt.show()
```

## 2. Phân tích hồi quy

```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

X = df.drop('loan_status', axis=1)  # Các biến độc lập
y = df['loan_status']

# Chia dữ liệu thành tập huấn luyện và tập kiểm thử
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Xây dựng mô hình hồi quy logistic
log_reg = sm.Logit(y_train, X_train).fit()

# In kết quả
print(log_reg.summary())
```
