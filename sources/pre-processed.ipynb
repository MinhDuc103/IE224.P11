```python
#Đọc dữ liệu từ file csv
df=pd.read_csv('credit_risk_dataset.csv')
df
```

### Xử lý dữ liệu

```python
#Tiến hành kiểm tra các giá trị phân biệt với biến person_age
#Phát hiện có giá trị bất thường (123,144) trong biến
age_unique=df['person_age'].unique()
age_unique
```

```python
#Tiến hành kiểm tra các giá trị phân biệt của biến person_emp_length
#Phát hiện có giá trị bất thường (123) trong biến
person_emp_length_unique=df['person_emp_length'].unique()
person_emp_length_unique
```

```python
#Tiến hành xuất các dòng có chứa giá trị tuổi là 123
df[df['person_age']==123]
```

```python
#Tiến hành xuất các dòng có chứa giá trị tuổi là 144
df[df['person_age']==144]
```

```python
#Tiến hành xuất các dòng có chứa giá trị thời gian làm việc là 123
df[df['person_emp_length']==123]
```

```python
#Các giá trị trên có ảnh hưởng tới cột dữ liệu nên tiến hành loại bỏ các dòng chứa giá trị trên.
df=df[df['person_age'] !=123]
df=df[df['person_age'] !=144]
df=df[df['person_emp_length'] !=123]
df
```

```python
#Tiến hành điền khuyết các giá trị trong biến loan_int_rate
from sklearn.impute import SimpleImputer

# Khởi tạo imputer với chiến lược trung vị
imputer = SimpleImputer(strategy='median')

# Điền giá trị khuyết cho cột 'loan_int_rate'
df[['loan_int_rate']] = imputer.fit_transform(df[['loan_int_rate']])
df['loan_int_rate']
```

```python
#Tiến hành điền khuyết các giá trị trong biến person_emp_length
from sklearn.impute import SimpleImputer

# Khởi tạo imputer với chiến lược trung vị
imputer = SimpleImputer(strategy='median')

# Điền giá trị khuyết cho cột 'loan_int_rate'
df[['person_emp_length']] = imputer.fit_transform(df[['person_emp_length']])
df['person_emp_length']
```

```python
#Nhận thấy các cột thu nhập (person_income), khoản vay(loan_amnt), tỷ lệ(loan_perscentr_income) có đơn vị đo lường khác nhau cần được đưa về cùng một thang đo để các mô hình học máy hoạt động hiệu quả hơn.
#Tránh chi phối bởi các biến lớn: Nếu không chuẩn hóa, các biến có giá trị lớn (như thu nhập) sẽ lấn át các biến khác trong quá trình huấn luyện mô hình.
#Tiến hành chuẩn hóa min-max scaling để chuẩn hóa các cột 'person_income', 'loan_amnt', 'loan_percent_income'
from sklearn.preprocessing import MinMaxScaler

# Lấy các cột cần chuẩn hóa
cols_to_scale = ['person_income', 'loan_amnt', 'loan_percent_income','person_age','person_emp_length','loan_int_rate','cb_person_cred_hist_length']

# Tạo scaler
scaler = MinMaxScaler()

# Thực hiện chuẩn hóa
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

print("Dữ liệu sau khi chuẩn hóa bằng Min-Max Scaling:")
print(df[cols_to_scale].head())
```

```python
# Áp dụng One-Hot Encoding trực tiếp
df = pd.get_dummies(df, columns=['person_home_ownership', 'loan_intent', 'cb_person_default_on_file'], drop_first=False)
print(df.head())
```

```python
# Chuyển đổi toàn bộ dataframe từ True/False sang 0/1
df = df.replace({True: 1, False: 0})
```

```python
#Đối với biến 'loan_grade' là biến phân loại theo thứ tự, tiến hành sử dụng rank-based scaling để chuẩn hóa dữ liệu biến trên
#Kiểm tra biến loan_grade trong dataset
if 'loan_grade' in df.columns:
    #Tiến hành rank-based scaling, khi mà các giá trị được chuẩn hóa về từ 0 đến 1 và được đánh số theo thứ tự tăng dần
    df['loan_grade_rank'] = df['loan_grade'].rank(method='dense').astype(float)
    df['loan_grade_scaled'] = df['loan_grade_rank'] / df['loan_grade_rank'].max()
else:
    print("'loan_grade' column not found in the dataset.")
df
```
